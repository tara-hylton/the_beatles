```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
beatles_df = pd.read_csv('raw_data/the_beatles_songs_dataset.csv')
beatles_df
```

Removing covers of other bands songs to work with only origin Beatles songs

```python
beatles_df = beatles_df[beatles_df.Cover != 'Y']
beatles_df = beatles_df.drop(['Popularity', 'Duration', 'Key', 'Mode','Tempo', 'Time_signature', 'Valence', 'Danceability', 'Energy', 'Loudness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Speechiness'], axis=1)
beatles_df = beatles_df.drop(['URI','Single certification', 'Genre', 'Styles', 'Moods', 'Other releases', 'Cover'], axis=1)
beatles_df
```

Sorting songs that hit number 1 on the Billboard charts

```python
beatles_df_hits = beatles_df.copy()
column_to_clean_2 = 'Weeks at No1 (Billboard)'
beatles_df_hits = beatles_df_hits.dropna(subset=column_to_clean_2)
beatles_df_hits
```

Showing how long songs remained at number one on the Billboard charts

```python
sns.barplot(x='Title', y='Weeks at No1 (Billboard)', data=beatles_df_hits)
plt.ylabel='Weeks on chart'
plt.xlabel=' '
plt.title("Beatles' Number One Hits")
plt.xticks(rotation=90)
plt.show()
```

Sorting songs by writer

```python
beatles_df_writers =beatles_df.copy()
column_to_clean_3 = 'Songwriter(s)'
beatles_df_writers = beatles_df_writers.dropna(subset=column_to_clean_3)
beatles_df_writers
```

Showing trends by writer and year

```python

sns.scatterplot(x='Year', y='Songwriter(s)', data=beatles_df_writers)
```

```python
sns.lineplot(x='Year', y='Songwriter(s)', data=beatles_df_writers)
```
